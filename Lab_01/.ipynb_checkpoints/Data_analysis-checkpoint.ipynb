## Exercise 1.1
### Estimation of $\langle r \rangle$ and $\langle \sigma^2 \rangle$ using the blocking method

I generated $M=10^4$ extractions using the Pseudo-Random Number generator downloaded from the Ariel website of the course, collecting the set  $\{r_i \}$, in order to estimate:

$$ \langle r \rangle = \int_0^1 r\,\mathrm{d}r = 1/2 \qquad \mathrm{and} \qquad \sigma^2 = \int_0^1 (r-1/2)^2 dr = 1/12. $$

For computing the statistical uncertaines of the process i implemented the **blocking method**.

Note that the variance for an event A is given by:

$$ \sigma_A^2 \equiv \langle (A-\langle A \rangle)^2 \rangle = \langle A^2 \rangle -  \langle A \rangle^2.$$ 

Our choice for computing the uncertain is based on the idea of dividing the entire set of $M$ events in $N$ blocks, with $N$ submultiple of $M$. Therefore, i compute che average value of A in each block, obtaining the set : $\{\langle A_i \rangle\}_{i=1}^{N}$. In the same time i obtain also the same set of values for $A^2$.  The desired estimators are:

 $$ \langle A^2 \rangle \simeq \frac{1}{N} \sum_{i=1}^N A^2_i \quad \quad \langle A\rangle^2 \simeq \left( \frac{1}{N} \sum_{i=1}^N A_i \right)^2, $$
 
 and, finally, the standard deviation is:
 
 $$\frac{\sigma}{\sqrt{N-1}} \simeq \sqrt{\frac{1}{N-1} \left[ \frac{1}{N} \sum_{i=1}^N A^2_i - \left( \frac{1}{N} \sum_{i=1}^N A_i \right)^2 \right]}. $$
 
 I used this method for compute che uncertain in the estimation of $\langle r \rangle$ and $\langle \sigma^2 \rangle$.   By choosing $N=10^2$ i obtained the results reported in the following graphics, in which the estimated value is represented by the thick black line, with the correspondent errors (vertical grey lines). I also tracked the orrizzontal red line corresponding to the theoretical value.

```python
x = np.arange(100)

plt.figure(figsize=(16,5))

plt.subplot(1,2,1)
plt.title("Expected value for a uniform random sample (M = 1e4, N=100).")
plt.errorbar(x, uniform.T[0]-0.5, yerr=uniform.T[1], color='black', alpha=0.5, label='Trend', linewidth=2)
plt.hlines(0,0,101, color='red', linestyle='--', label="Theoric expected value", linewidth=3, alpha=0.6)
plt.ylabel("<r>-0.5")
plt.xlabel("#throws")
plt.legend()

plt.subplot(1,2,2)
plt.title("Expected value of the variance in a uniform random sample (M = 1e4, N=100).")
plt.errorbar(x, varianc.T[0]-(1/12), yerr=varianc.T[1], color='black', alpha=0.5, label='Trend', linewidth=2)
plt.hlines(0,0,101, color='red', linestyle='--', label="Theoric expected value", linewidth=3, alpha=0.6)
plt.ylabel("<(r-0.5)**2>-1/12")
plt.xlabel("#throws")
plt.legend()

plt.show()
```

### Evaluation of $\chi^2$

In this second part, we decided to divide the range [0,1] of the random numbers in $N=10^2$ intervals of lenght $\ell = 1/N$. The target is to evaluate how many extractions $n_i$ fall into the $[x_i,x_i+\ell]$ interval. Considering a set of $M=10^4$ extractions, the expected frequency for each interval is $n=M/N=100.$ To estimate the error between the observations and the expected values we used the Pearson Cumulative test statistic $\chi^2$, which in this case is:

$$\chi^2 = \sum_{i=1}^M \frac{\left( n_i - n/M \right)^2}{n/M}.$$

I repeated the evaluation for $j=1,...,100$ times and computed the final value of each one in the following graphic.

```python
plt.figure(figsize=(13,5))

plt.title('$\chi^2$ for i=1,...,100 samples')
plt.plot(chiquad, color='black', alpha=0.7, linestyle=' ', marker='.', markersize=10)
plt.plot(chiquad, color='black', alpha=0.7, linewidth=0.7)
plt.xlabel("#sample")
plt.ylabel("$\chi^2$")
plt.show()
```
