```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_theme(font_scale=1.7, style='whitegrid')
```

```python
ist_sol = np.loadtxt("Code/7.1/istant_solid.dat")
ist_liq = np.loadtxt("Code/7.1/istant_liquid.dat")
```

```python
plt.figure(figsize=(18,12))


plt.subplot(2,1,1)
plt.title('')
plt.plot(ist_sol.T[0], linewidth=3, alpha=0.5, color='orange')
plt.plot(ist_liq.T[0], linewidth=3, alpha=0.5, color='red')

plt.subplot(2,1,2)
plt.plot(ist_sol.T[1], linewidth=3, alpha=0.5, color='orange')
plt.plot(ist_liq.T[1], linewidth=3, alpha=0.5, color='red')


plt.show()
```
