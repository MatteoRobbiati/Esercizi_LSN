```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_theme(font_scale=1.7, style='whitegrid')
```

```python
istant = []

istant.append(np.loadtxt("Code/7.1/istant_solid.dat"))
istant.append(np.loadtxt("Code/7.1/istant_liquid.dat"))
istant.append(np.loadtxt("Code/7.1/istant_gas.dat"))

phase = ['Solido', 'Liquido', "Gas"]
colors = ['orange', 'red', 'purple']
```

```python
plt.figure(figsize=(25,20))

for i in range(3):
    plt.subplot(3,1,i+1)
    plt.title(phase[i])
    plt.plot(istant[i].T[0], linewidth=3, alpha=0.5, color='blue')
    plt.plot(istant[i].T[2], linewidth=3, alpha=0.5, color='red')
    plt.xlabel('# passi $\cdot 100$')
    plt.ylabel('Grandezze in LJ units')

plt.tight_layout()    
plt.show()
```
